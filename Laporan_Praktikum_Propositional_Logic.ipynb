Untuk mencari nilai **a2** dengan pernyataan “There is no pit in (2,2)”, simbol yang digunakan adalah:
```
B21, B12, P13, P22, P31
```
atau
```py
P13, P22, P31, B21, B12 = expr('P13, P22, P31, B21, B12')
```
<br>

Dan, dimana sentence yang akan kita gunakan yaitu:
```
R1: P(2,2) <=> (B(2,1) & B(1,2))
R2: P(3,1) <=> B(2,1)
R3: P(1,3) <=> B(1,2)
R4: -B(1,2)
R5: B(2,1)
```
atau
```py
wumpus_kb.tell(~P11)
wumpus_kb.tell(P22 | '<=>' | ((B21 & B12)))
wumpus_kb.tell(P31 | '<=>' | (B21))
wumpus_kb.tell(P13 | '<=>' | (B12))
wumpus_kb.tell(~B12)
wumpus_kb.tell(B21)
```
<br>

Dengan demikian, kode lengkapnya adalah sebagai berikut:



````python
wumpus_kb = PropKB()

P13, P22, P31, B21, B12 = expr('P13, P22, P31, B21, B12')

wumpus_kb.tell(~P11)
wumpus_kb.tell(P22 | '<=>' | ((B21 & B12)))
wumpus_kb.tell(P31 | '<=>' | (B21))
wumpus_kb.tell(P13 | '<=>' | (B12))
wumpus_kb.tell(~B12)
wumpus_kb.tell(B21)

wumpus_kb.ask_if_true(~P22)
````

Kode tersebut menghasilkan output 'True' atau "There is NO pit in (2,2)", hasil tersebut sesuai dengan hasil truth table manual.

````python
wumpus_kb = PropKB()

W13, W22, W31, S21, S12 = expr('W13, W22, W31, S21, S12')

wumpus_kb.tell(W22 | '<=>' | ((S21 & S12)))
wumpus_kb.tell(W31 | '<=>' | (S21))
wumpus_kb.tell(W13 | '<=>' | (S12))
wumpus_kb.tell(S12)
wumpus_kb.tell(~S21)

wumpus_kb.ask_if_true(W13)
````
